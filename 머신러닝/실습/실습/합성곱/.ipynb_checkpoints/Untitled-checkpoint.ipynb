```python
class ResidualUnit(keras.layers.Layer):
    def __init__(self,filters,strides=1,activition='relu',**kwargs):
        super().__init__(**kwargs)
        self.activition=karas.activition.get(activition)
        self.main.layers=[
            keras.layers.Conv2D(filters,3,strides=strides,padding="same",usebias=False),
            keras.layers.BatchNormalization(),
            self.activition,
            keras.layers.Conv2D(filters,3,strides=1,padding="same",usebias=False),
            keras.layers.BatchNormalization()
        ]
        self.skip_layers=[]
        if strides>1:
            self.skip_layers=[
                keras.layers.Conv2D(filters,1,strides=strides,padding="same",usebias=False),
                keras.layers.BatchNormalization()
            ]
            
    def call(self,inputs):
        z=inputs
        for layer in self.main_layers:
            z=layer(z)
        skip_z=inputs
        for layer in self.skip_layers:
            skip_z=layer(skip_z)
        return self.activition(z+skip_z)
```

```python
import tensorflow_datasets as tfds
```

```python
dataset,info=tfds.load('tf_flowers',as_supervised=True,with_info=True)
```

```python
datasetsize=info.splits['train'].num_examples
classname=info.features['label'].names
nclass=info.features['label'].num_classes
```

```python

test,valid,train=tfds.load('tf_flowers',split=["train[:10%]","train[10%:25%]","train[25%:]"],as_supervised=True)
```

```python
plt.figure(figsize=(12, 10))
index = 0
for image, label in train.take(3):
    index += 1
    plt.subplot(3, 3, index)
    plt.imshow(image)
    plt.title("Class: {}".format(classname[label]))
    plt.axis("off")

plt.show()
```

```python
def preprocess(image,label):
    resize=tf.image.resize(image,[224,224])
    final=keras.applications.xception.preprocess_input(resize)
    return final,label
```

```python
batchsize=32
train=train.shuffle(1000)
train=train.map(preprocess).batch(batchsize).prefetch(1)
valid=valid.map(preprocess).batch(batchsize).prefetch(1)
test=test.map(preprocess).batch(batchsize).prefetch(1)
```

```python
plt.figure(figsize=(12, 12))
for X_batch, y_batch in train.take(1):
    for index in range(9):
        plt.subplot(3, 3, index + 1)
        plt.imshow(X_batch[index] / 2 + 0.5)
        plt.title("Class: {}".format(classname[y_batch[index]]))
        plt.axis("off")

plt.show()
```

```python
base=keras.applications.xception.Xception(weights='imagenet',include_top=False)
avg=keras.layers.GlobalAveragePooling2D()(base.output)
output=keras.layers.Dense(nclass,activation='softmax')(avg)
model=keras.Model(inputs=base.input,outputs=output)
```

```python
for layer in base.layers:
    layer.trainable=False
```

```python
optimizer=keras.optimizers.SGD(lr=0.2,momentum=0.9,decay=0.01)
model.compile(loss='sparse_categorical_crossentropy',optimizer=optimizer,metrics=['accuracy'])
history=model.fit(train,epochs=5,validation_data=valid)
```

```python
for layer in base.layers:
    layer.trainable=True
optimizer=keras.optimizers.SGD(lr=0.01,momentum=0.9,decay=0.001)
model.compile(loss='sparse_categorical_crossentropy',optimizer=optimizer,metrics=['accuracy'])
#history=model.fit(train,epochs=5,validation_data=valid)
```

```python
base=keras.applications.xception.Xception(weights='imagenet',include_top=False)
avg=keras.layers.GlobalAveragePooling2D()(base.output)
classoutput=keras.layers.Dense(nclass,activation='softmax')(avg)
loc_output=keras.layers.Dense(4)(avg)
model=keras.Model(inputs=base.input,outputs=[classoutput,loc_output])
model.compile(loss=["sparse_categorical_crossentropy",'mse'],loss_weights=[0.8,0.2],optimizer=optimizer,
             metrics=['accuracy'])
```
